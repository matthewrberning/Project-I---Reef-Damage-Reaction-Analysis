```python
# import dependancies
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
#pull in google trend data 
google_trend_damage_df = pd.read_csv("C:/Users/OI/Desktop/g_trend.csv")
gt_coral_df = pd.read_csv("C:/Users/OI/Desktop/coral.csv")
gt_reef_df = pd.read_csv("C:/Users/OI/Desktop/reef.csv")
gt_coralreef_df = pd.read_csv("C:/Users/OI/Desktop/coral_reef.csv")
gt_reefdiving_df = pd.read_csv("C:/Users/OI/Desktop/reef_diving.csv")
gt_greatbarrier_df = pd.read_csv("C:/Users/OI/Desktop/great_barrier.csv")
gt_reef_scuba_df = pd.read_csv("C:/Users/OI/Desktop/reef_scuba.csv")

gt_coral_df.head(1)
```

```python
google_trend_all_df = pd.merge(google_trend_damage_df, gt_coral_df, on='Month')
google_trend_all_df = pd.merge(google_trend_all_df, gt_reef_df, on='Month')
google_trend_all_df = pd.merge(google_trend_all_df, gt_coralreef_df, on='Month')
google_trend_all_df = pd.merge(google_trend_all_df, gt_reefdiving_df, on='Month')
google_trend_all_df = pd.merge(google_trend_all_df, gt_greatbarrier_df, on='Month')
google_trend_all_df = pd.merge(google_trend_all_df, gt_reef_scuba_df, on='Month')


google_trend_all_df.head(100)
```

```python
google_trend_all_df = google_trend_all_df.rename(columns={"coral: (Worldwide)": "coral","reef: (Worldwide)":"reef", "coral reef: (Worldwide)":"coral reef" })

google_trend_all_df.head()
```

```python
google_trend_all_df["average"] = google_trend_all_df.mean(axis=1)

google_trend_all_df = google_trend_all_df.set_index("Month")

google_trend_all_df.head(100)
```
